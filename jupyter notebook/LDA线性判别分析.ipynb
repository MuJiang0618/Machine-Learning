# LDA线性判别分析

性能不够好且不稳定,因为训练集太小,只有299个样本,每个样本却有22个特征,<机器学习实战>用logistic回归得到的性能也是这样

```python
import numpy as np
from sklearn.cross_validation import train_test_split
import pandas as pd
```

```python
dataSet = pd.read_table(r'D:\dataSet\HorseColicTraining.txt', header=None)
dataSet.head()
```

```python
trainSet, testSet = train_test_split(dataSet, test_size=0.1)
```

```python
def getW(dataSet):
#     trainSet, testSet = train_test_split(dataSet, test_size=0.1)
    trainSet_0 = trainSet[trainSet[21] == 0] ; trainSet_1 = trainSet[trainSet[21] == 1]
    trainSet_0_X = trainSet_0[[i for i in range(0, 21)]] ; train_0_Y = trainSet_0[21]
    trainSet_1_X = trainSet_1[[i for i in range(0, 21)]] ; train_1_Y = trainSet_1[21]
    miu0 = trainSet_0_X.mean(axis=0) ; miu1 = trainSet_1_X.mean(axis=0)
    meaned_0, meaned_1 = trainSet_0_X - miu0, trainSet_1_X - miu1
    Sw = np.trace(np.dot(meaned_0, meaned_0.T)) + np.trace(np.dot(meaned_1, meaned_1.T))
    w = np.array(1/Sw * (miu0 - miu1))
    w = w.reshape(1, w.shape[0])
    #注意miu是投影后的样本均值点
    miu0 = np.multiply(trainSet_0_X, w).mean(axis=0)
    miu1 = np.multiply(trainSet_1_X, w).mean(axis=0)
    
    return w, miu0, miu1

def predict(w, miu0, miu1):
    test_X = testSet[[i for i in range(0, 21)]]
    test_Y = testSet[21]

    score = np.multiply(test_X, w)
    #注意要用绝对距离
    predict_result = abs((score - miu1).sum(axis=1)) < abs((score - miu0).sum(axis=1))
    precision = sum(predict_result == test_Y) / len(testSet)
    
    return precision

def main():
    w, miu0, miu1 = getW(dataSet)
    precision = predict(w, miu0, miu1)
    print('precision: %f' %precision)
    
    return

main()
```
