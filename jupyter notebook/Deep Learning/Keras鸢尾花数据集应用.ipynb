# Keras鸢尾花分类

## 经验
性能不好时,尝试改变optimizer, 有时SGD效果很好(但是训练得到的性能具有随机性, 可能要fit()多次); 如果出现振荡, 提高epochs, 降低lr

如何获取输出层输出的值?

```python
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from keras.models import Sequential, Model  #Sequential:一层一层地加神经网络
from keras.layers import Dense, Input    #Dense:全连接神经网络
from keras.utils import np_utils
import numpy as np
import tensorflow as tf
from keras.optimizers import SGD
```

```python
X = load_iris().data ; y = load_iris().target
y_oneHot = np_utils.to_categorical(y, num_classes=3)
train_x, test_x, train_y, test_y = train_test_split(X, y_oneHot, test_size=0.8)
```

## 搭建模型

```python
model = Sequential([ Dense(4, input_dim=4, activation='relu'),
                    Dense(6, input_dim=6, activation='relu'),
                     Dense(3, activation='softmax')])

model.compile(loss='categorical_crossentropy', optimizer=SGD(0.1), metrics=['accuracy'])
model.fit(train_x, train_y, epochs=300)
```

 因为是SGD, 所以模型的性能具有随机性

```python
model.evaluate(test_x, test_y)
```
